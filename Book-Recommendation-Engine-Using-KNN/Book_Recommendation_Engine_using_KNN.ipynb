# Preparing the Dataset for **K-Nearest Neighbors** recommendation model

````python
# Creating a new 'Pandas Dataframe' from the remaining rating(s)
df = df_ratings_rm.pivot_table(index=['user'],columns=['isbn'],values='rating').fillna(0).T

# Getting the first 5 entries of this Dataframe
df.head()
````

````python
# Join the books and ratings Dataframes to get the title of the books
df.index = df.join(df_books.set_index('isbn'))['title']

# Sorting the index of the Dataframe
df = df.sort_index()

# Getting the first 5 entries of the sorted Dataframe
df.head()
````

````python
# Locate the first five entries of a book
df.loc["The Queen of the Damned (Vampire Chronicles (Paperback))"][:5]
````

# Creating the model

````python
# Initializing the KNN model with 'cosine' metric
model = NearestNeighbors(metric='cosine')

# Training the model
model.fit(df.values)
````

#Getting the recommended books

### Creating a function to get recommended books

````python
# function to return recommended books - this will be tested
def get_recommends(book_title = ""):
  try: # to find the required book in the Dataframe
    book = df.loc[book_title]
  except KeyError as e:
    print('The given book', e, 'does not exist')
    return

  # Set number of neighbours to 6
  num_neighbours = 6
  # Get the distance and indexes of the recommended books
  distance, indexes = model.kneighbors(
      [book.values],
      n_neighbors=num_neighbours
  )

  # Get the values of the (sorted) Dataframe of recommended books
  recommended_books = pd.DataFrame({
      'title'   : df.iloc[indexes[0]].index.values,
      'distance': distance[0]
    }) \
    .sort_values(by='distance', ascending=False) \
    .head(5).values

  # Return a list including the title of the book and Dataframe of recommended books
  return [book_title, recommended_books]
````

````python
[book_title, recommended_books] = get_recommends("Where the Heart Is (Oprah's Book Club (Paperback))")
print(type(recommended_books))
````
